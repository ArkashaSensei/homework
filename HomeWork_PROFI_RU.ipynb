# 1. Написать функцию на Python

```python
a = [1, 3, 2, 1, 5, 3, 5, 1, 4]
def Sas(a):
    d = dict()
    for i in a:
        d[i] = 1
    return list(d.keys())
Sas(a)
```

Асимптотическая сложность o(N), так как в списке N элементов, а запись в словарь имеет сложность o(1).
